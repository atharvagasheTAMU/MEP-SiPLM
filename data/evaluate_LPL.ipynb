```python
import pandas as pd
```

```python
result_path = "../supervised/CAGI_LPL_ESM2.csv"
reference_file = "/home/yxliu/Documents/CAGI7/dataset/LPL/LPL_varsOnly.csv"
result = pd.read_csv(result_path)
ref = pd.read_csv(reference_file)
print(len(result),len(ref))
```

```python
scores = result["Score"].tolist()
mutant = ref["hgvs_pro"].tolist()
mutant = ['WT'] + mutant
zipped_data = list(zip(mutant, scores))

# Define the column names
columns = ['Mutant', 'Scores']

# Create the DataFrame
df = pd.DataFrame(zipped_data, columns=columns)
df.head()
```

```python
df_no_Ter = df[~df["Mutant"].str.contains("Ter")]
```

```python
scores = df["Scores"].tolist()
scores
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Create your list of data


# For a more robust example with more data points:
# data = np.random.randn(500) * 15 + 100 # Normally distributed data

# 2. Plot the histogram
plt.hist(scores, bins=100, rwidth=0.9, color='skyblue', edgecolor='black')
line_value = scores[0]
plt.axvline(x=line_value, color='red', linestyle='--', linewidth=2, label=f'Threshold at {line_value}')
# 3. Add labels and title
plt.title('Simple Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.savefig("LPL_distribution.png")
# 4. Display the plot
plt.show()
```
